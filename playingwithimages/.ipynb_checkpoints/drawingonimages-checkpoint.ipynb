```python
import cv2
import numpy as np


image = np.zeros((512,512,3), np.uint8)


cv2.imshow("Black box",image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
Drawing a circle line in the square
```

```python
import cv2
import numpy as np


image = np.zeros((512,512,3), np.uint8)

cv2.line(image, (0,0),(513,513), (128,128,45), 5)

cv2.imshow("Weird line",image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
Drawing a rectangle 
```

```python
import cv2
import numpy as np


image = np.zeros((512,512,3), np.uint8)

cv2.rectangle(image, (150,150),(300,300), (128,128,45), -1)

cv2.imshow("Weird line",image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
