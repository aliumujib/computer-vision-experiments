```python
import cv2
import numpy as np

image = cv2.imread("ALIU_ABDULMUJEEB_OLOLADE.jpg")
cv2.imshow("Input Image", image)
cv2.waitKey(0)

height, width = image.shape[:2]

sobel_x = cv2.Sobel(image, cv2.CV_64F, 0,1, ksize= 5)
sobel_y = cv2.Sobel(image, cv2.CV_64F, 1,0, ksize= 5)

cv2.waitKey(0)

#convert to grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

sobel_OR = cv2.bitwise_or(sobel_x, sobel_y)
cv2.imshow("Sobel X", sobel_x)
cv2.waitKey(0)
cv2.imshow("Sobel Y", sobel_y)
cv2.waitKey(0)
cv2.imshow("Sobel OR", sobel_OR)
cv2.waitKey(0)

laplacian = cv2.Laplacian(image, 40,250)
cv2.imshow("Laplacian Image", laplacian)
cv2.waitKey(0)

#find canny edges, helps reduce noise
canny_edge = cv2.Canny(gray_image, 40,250)
cv2.imshow("Canny Image", canny_edge)
cv2.waitKey(0)


cv2.destroyAllWindows()
```
